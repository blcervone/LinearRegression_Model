Create a linear regression model to predict a student’s college GPA using their high school SAT test scores.

```python
import numpy as no
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels

%matplotlib inline
```

```python
df = pd.read_csv("./sat_gpa_v1.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df['sat_sum'].values.reshape(-1,1)
y = df['fy_gpa']

model = LinearRegression()
model.fit(X,y)
```

```python
print(model.intercept_, model.coef_, model.score(X, y))
```

```python
import statsmodels.api as sm
```

```python
results = sm.OLS(y, X).fit()
```

```python
print(results.summary())
```

```python
plt.figure(figsize = (20.0,10.0))
lin_plt = sns.regplot(data = df, x = 'sat_sum', y= 'fy_gpa', scatter_kws = {'s': 100, 'alpha': 0.4})
sns.despine()
plt.title("Correlation between First-Year College GPA and SAT Scores", fontsize = 18)
plt.xlabel("SAT Score (Math + Verbal)", fontsize = 14)
plt.ylabel("First-Year GPA", fontsize = 14)
plt.tick_params(axis='both', which='major', labelsize=12)
```
